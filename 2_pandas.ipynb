```python
import pandas as pd
import numpy as np
x = pd.Series(np.random.rand(4))
x
```

```python
# Indexing and slicing
print(x[3])
print(x[1:3])
```

```python
print(x.index)
x.index = ['a','b','c','d']
x
```

```python
print(x['a':'c'])
print(x[0:4:2])
```

```python
# Creating two series
numbers1 = pd.Series(np.random.rand(5))
numbers2 = pd.Series(np.random.rand(5))
print(numbers1,numbers2)
```

```python
# Creating dataframe with series
df = pd.DataFrame([numbers1,numbers2])
df
```

```python
# Transpose - Converting observations(rows) into features(column)
df = df.T
df
```

```python
# Changing the column name
df.columns = ['S1','S2']
df
```

```python
# Creating dictionary
details = {'Name':["John","Smith","Kamal","Mary"],
         'Gender':["M","M","M","F"]}
details
```

```python
df = pd.DataFrame(details)
df
```

```python
# Adding a column
df['Company'] = ["Amazon","TCS","Apple","Microsoft"]
df["Age"] = [np.nan,33,25,25]
df
```

```python
df['Agee'] = [50,60,45,58]
df
```

```python
# Removing unwanted column
df = df.drop(['Agee'],axis=1)
df
```

```python
# Location (loc)
# Give the actual row and column name as parameter
df.loc[1:3,"Name":"Company"]
```

```python
# Index Location (iloc)
# Give the actual row and column index value as the parameter
df.iloc[1:3,2:3]
```

```python
# Adding a row
df.loc[4] = ["Sam","F","Google",29]
df
```

```python
# Removing a row
df.drop([3],axis=0)
```

```python
# Finding Null values
print(df.isnull())
df.isnull().sum()
```
