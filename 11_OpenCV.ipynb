## Face detection

```python
img = cv2.imread('/content/face.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
haarCas = '/content/haarcascade_frontalface_alt (1).xml'
haarCas
```

```python
faces = cv2.CascadeClassifier(haarCas)
faces
```

```python
face = faces.detectMultiScale(img, scaleFactor=1.1, minNeighbors = 4)
face
```

```python
for (x,y,w,h) in face:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0,3))
plt.imshow(img)
```

## Number plate detection

```python
img = cv2.imread('/content/NumberPlate.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
haarcas = '/content/haarcascade_russian_plate_number (1).xml'
haarcas
```

```python
number_plates = cv2.CascadeClassifier(haarcas)
number_plates
```

```python
number_plate = number_plates.detectMultiScale(img,scaleFactor=1.1, minNeighbors = 4)
number_plate
```

```python
for (x,y,w,h) in number_plate:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),7)
plt.imshow(img)
```
